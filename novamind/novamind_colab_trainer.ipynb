### 1. The Genuine PyTorch Predictive Substrate

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from datasets import load_dataset
import time

class DynamicPredictiveNetwork(nn.Module):
    def __init__(self, initial_concepts=64, embedding_dim=128):
        super().__init__()
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        print(f"🔥 Motor Rodando em: {self.device.type.upper()}")
        
        self.vocab_size = initial_concepts
        self.embedding_dim = embedding_dim
        self.embeddings = nn.Parameter(torch.randn(self.vocab_size, self.embedding_dim, device=self.device) * 0.02)
        self.transition_weights = nn.Parameter(torch.randn(self.embedding_dim, self.embedding_dim, device=self.device) * 0.02)
        self.lr = 0.01
        self.neurogenesis_threshold = 0.95 # Só cria neurônio se for realmente surpreso (altamente alienígena)
        self.token_to_id = {}
        self.id_to_token = {}
        
    def get_token_id(self, token):
        if token not in self.token_to_id:
            if len(self.token_to_id) >= self.vocab_size:
                 self._spawn_neuron()
            new_id = len(self.token_to_id)
            self.token_to_id[token] = new_id
            self.id_to_token[new_id] = token
        return self.token_to_id[token]

    def forward(self, concept_indices):
        x = self.embeddings[concept_indices]
        return F.silu(torch.matmul(x, self.transition_weights))

    def _spawn_neuron(self):
        with torch.no_grad():
            new_embedding = torch.randn(1, self.embedding_dim, device=self.device) * 0.02
            self.embeddings = nn.Parameter(torch.cat([self.embeddings, new_embedding], dim=0))
            self.vocab_size += 1
            if self.vocab_size % 1000 == 0:
                 print(f"\n🧬 EXPLOSÃO DE NEUROGÊNESE: Cérebro físico expandiu para {self.vocab_size} conceitos geométricos.")

    def train_step(self, text_chunk):
        # Simple heuristic subword chunking for real scale (groups of ~4 chars to simulate emergent concepts)
        tokens = [text_chunk[i:i+4] for i in range(0, len(text_chunk), 4)]
        if len(tokens) < 2: return 0.0
        
        indices = [self.get_token_id(t) for t in tokens]
        
        inp = torch.tensor(indices[:-1], dtype=torch.long, device=self.device)
        tgt = torch.tensor(indices[1:], dtype=torch.long, device=self.device)
        
        predictions = self.forward(inp)
        targets = self.embeddings[tgt].detach()
        
        surprise = 1.0 - F.cosine_similarity(predictions, targets, dim=-1)
        loss = surprise.mean()
        
        grad_embed = torch.autograd.grad(loss, self.embeddings, retain_graph=True)[0]
        grad_transition = torch.autograd.grad(loss, self.transition_weights)[0]
        
        with torch.no_grad():
            self.embeddings -= self.lr * grad_embed
            self.transition_weights -= self.lr * grad_transition
            
        return loss.item()

    def interact(self, prompt, max_responses=20):
        """ A Mente respondendo usando a topologia treinada. """
        print(f"\n👤 Humano: {prompt}", end="")
        print("\n🧠 NovaMind: ", end="")
        
        # Trata o input mais recente como contexto inicial
        prompt_tokens = [prompt[i:i+4] for i in range(0, len(prompt), 4)]
        current_token = prompt_tokens[-1] if prompt_tokens else " "
        
        with torch.no_grad():
            for _ in range(max_responses):
                idx = self.get_token_id(current_token)
                inp = torch.tensor([idx], dtype=torch.long, device=self.device)
                
                # Prediz a próxima topologia
                predicted_topology = self.forward(inp)
                
                # Acha o conceito existente mais próximo dessa geometria esperada (Cosine Similarity)
                similarities = F.cosine_similarity(predicted_topology, self.embeddings, dim=-1)
                best_idx = similarities.argmax().item()
                
                next_token = self.id_to_token[best_idx]
                print(next_token, end="", flush=True)
                current_token = next_token
                time.sleep(0.05) # Para dar o efeito de pensar
        print("\n")
        
mind = DynamicPredictiveNetwork(initial_concepts=10, embedding_dim=128)
```

### 2. Massive Dataset Ingestion (The World Stream)
Here we connect the Mind to a massive stream of human knowledge (WikiText-2, part of Wikipedia). Notice how it will spawn thousands of conceptual geometry axes.

```python
print("📡 Conectando ao fluxo global de conhecimento (WikiText dataset)...")
dataset = load_dataset("wikitext", "wikitext-2-v1", split="train")

print("\n🚀 INICIANDO TREINAMENTO TOPOLÓGICO MASSIVO...")
total_surprise = 0.0

# Lê as primeiras 500 grandes passagens de texto da wikipedia
for step, row in enumerate(dataset.select(range(500))):
    text = row['text'].strip()
    if len(text) < 50: continue # Pula linhas quase vazias
    
    surprise = mind.train_step(text)
    total_surprise += surprise
    
    if step % 20 == 0:
        avg_surprise = total_surprise / 20
        print(f"-> Absorvido artigo {step} | Nodos Ativos: {mind.vocab_size} | Tensão Neural (Surpresa): {avg_surprise:.4f}")
        total_surprise = 0.0

print("\n✅ BASE DE CONHECIMENTO PRIMÁRIA CONSOLIDADA.")
```
